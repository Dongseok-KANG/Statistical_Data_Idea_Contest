```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import Imputer
```

```python
train = pd.read_csv('기술통계19.csv',index_col='global_id')
service = pd.read_csv('실태조사_서비스업19.csv',index_col='global_id')
```

```python
line3=[]
for i in service.index:
    if i in train.index:
        line3.append(i)
    else:
        line3.append(0)
        
service_test=service[service.index == line3]

service_test['test'] = 1
```

```python
line4=[]
for i in service.index:
    if not i in train.index:
        line4.append(i)
    else:
        line4.append(0)
        
service_train=service[service.index == line4]

service_train['test'] = 0
```

```python
service = pd.concat([service_train, service_test])
```

손익계산서 변수들만 추출

```python
cols = [c for c in service.columns if c[:4] == 'INC_']
service=service[cols]
```

'-', 콤마 제거

```python
for i in cols:
    service[i] = service[i].replace(' - ',np.nan)
    service[i] = service[i].replace(',','')    
```

object 타입 float으로 변경

```python
for i in cols:
    service = service.astype({i: float})
```
