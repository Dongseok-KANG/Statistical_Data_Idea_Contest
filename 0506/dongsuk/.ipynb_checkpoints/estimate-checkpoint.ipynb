```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('기술통계19.csv',index_col='global_id')
service = pd.read_csv('실태조사_서비스업19.csv',index_col='global_id')
product = pd.read_csv('실태조사_제조업19.csv',index_col='global_id')
```

```python
line3=[]
for i in service.index:
    if i in train.index:
        line3.append(i)
    else:
        line3.append(0)
        
service_test=service[service.index == line3]

service_test['test'] = 1
```

```python
line4=[]
for i in service.index:
    if not i in train.index:
        line4.append(i)
    else:
        line4.append(0)
        
service_train=service[service.index == line4]

service_train['test'] = 0
```

```python
service = pd.concat([service_train, service_test])
```

```python
service
```

```python
line5=[]
for i in product.index:
    if i in train.index:
        line5.append(i)
    else:
        line5.append(0)
        
product_test=product[product.index == line5]

product_test['test'] = 1
```

```python
line6=[]
for i in product.index:
    if not i in train.index:
        line6.append(i)
    else:
        line6.append(0)
        
product_train=product[product.index == line6]

product_train['test'] = 0
```

```python
product = pd.concat([product_train, product_test])
```
